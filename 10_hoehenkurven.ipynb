### Höhenkurven

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Ausgangsgleichung
def V(T_0, p):
    R = 8.31446261815324
    result = R * T_0 / p
    return result

#Intervalle für T und p
T_values = [10, 50, 100, 200, 300]
p_values = range(1, 101)
T_values_result = {}

#Ergebnisse von V(T,p) in ein Dictionary speichern
for T_value in T_values:
    current_result = []
    for p_value in p_values:
        current_result.append(V(T_value, p_value))
    T_values_result[T_value] = current_result
```

```python
import seaborn as sns

#Plot erstellen
for T_value in T_values_result:
    sns.lineplot(x = p_values, y = T_values_result[T_value], label = f'T = {T_value}')

sns.set_style('whitegrid')  # Apply seaborn light blue theme

colors = ["#395a7f", "#6e9fc1", "#a3cae9", "#e9ecee", "#acacac"] #set custom color palette
sns.set_palette(palette = colors)

plt.grid(True)
plt.xlim(0, 100)
plt.ylim(0, 100)
plt.xlabel('Druck p')
plt.ylabel('Volumen V')
plt.title('Volumen V in Abhängigkeit von Druck p (R = 8.314)')
```
